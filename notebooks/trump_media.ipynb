```python
import yfinance as yf
import matplotlib.pyplot as plt
import seaborn as sns
import time

def fetch_and_plot(ticker):
    sns.set(style='whitegrid')
    plt.ion()  # Turn on interactive mode
    while True:
        # Fetch the latest hourly data
        data = yf.download(ticker, period="1d", interval="5h")

        # Clear the current plot
        plt.clf()

        # Plot the new data
        sns.lineplot(data=data['Close'], label='Hourly Close')
        plt.title(f'Hourly Close Price of {ticker}')
        plt.xlabel('Time')
        plt.ylabel('Close Price')
        plt.legend()
        plt.pause(0.1)  # Pause briefly to allow the plot to update

        # Wait for one hour
        time.sleep(3600)

# Example usage
ticker = "DJT"
fetch_and_plot(ticker)
```
